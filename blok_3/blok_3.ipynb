```python
# Подготовка признаков (идемпотентно)
df['Марка'] = df['Марка + модель'].str.split().str[0]
_m2q = {'Январь': 1, 'Февраль': 1, 'Март': 1, 'Апрель': 2, 'Май': 2, 'Июнь': 2,
        'Июль': 3, 'Август': 3, 'Сентябрь': 3, 'Октябрь': 4, 'Ноябрь': 4, 'Декабрь': 4}
df['Квартал'] = df['Месяц'].map(_m2q)

print('Продажи по годам:')
print(df['Год'].value_counts().sort_index())
```

## Аналитическая записка для руководства

**Динамика рынка 2008–2011** (выборка 5999 регистраций):

| Год | Продажи | Δ к пред. году |
|---|---|---|
| 2008 | 863 | — |
| 2009 | 826 | −4% |
| 2010 | 1501 | +82% |
| 2011 | 2809 | +87% |

После стагнации 2008–2009 рынок резко вырос в 2010–2011 гг., с пиком в **IV квартале 2011** (у VW Q4-2011 = 184 против Q4-2010 = 95).

**Внешние факторы:**
- **Мировой финансовый кризис 2008–2009** — провал спроса в 2009 г.
- **Валютный кризис в Беларуси 2011 г.**: девальвация рубля почти в 3 раза. Обесценивающиеся рубли население и бизнес вкладывали в товары длительного пользования, в т.ч. авто → ажиотажный спрос и пик IV кв. 2011.
- **Таможенный союз (2010–2011)**: ожидание роста ввозных пошлин и утилизационного сбора подталкивало к покупкам «впрок».
- Доступность автокредитов до середины 2011 г., затем сворачивание из-за роста ставок.

**Внутренние факторы:**
- Развитие дилерских сетей и расширение бюджетного модельного ряда (Renault, Kia, Hyundai, Geely).
- Концентрация спроса в Минске (≈48% регистраций) — платёжеспособность столицы.
- Смещение спроса в бюджетные седаны (Polo Sedan, Sandero, Accent, Rio) и SUV.

**Выводы:**
- Рост 2010–2011 в значительной мере разогнан разовым кризисным ажиотажем 2011 г., а не устойчивым улучшением платёжеспособности.
- В 2012 г. вероятна коррекция вниз: падение реальных доходов после девальвации, рост пошлин/утильсбора, исчерпание отложенного спроса.
- Поэтому прогноз, экстраполирующий тренд 2008–2011, следует трактовать как **оптимистичный сценарий**; консервативный сценарий предполагает продажи ниже.

## 2. Поквартальный прогноз продаж VW на 2012–2013 (Holt-Winters)

Метод: экспоненциальное сглаживание (Holt-Winters) с **аддитивным трендом** и **сезонностью период = 4 квартала**, обучено на 16 кварталах 2008–2011.

⚠️ Важно: обучающий ряд включает кризисный ажиотаж 2011 г., поэтому модель экстраполирует высокий тренд. Прогноз следует трактовать как **оптимистичный сценарий** — фактические продажи 2012 г. в реальности были ниже из-за коррекции рынка после девальвации.

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Поквартальный ряд продаж всего модельного ряда VW (2008–2011)
vw = df[df['Марка'] == 'Volkswagen']
q = vw.groupby(['Год', 'Квартал']).size().sort_index()
idx = pd.PeriodIndex([f'{y}Q{qq}' for y, qq in q.index], freq='Q')
ts = pd.Series(q.values, index=idx, dtype=float)

# Holt-Winters: аддитивный тренд + сезонность (4 квартала)
model = ExponentialSmoothing(ts, trend='add', seasonal='add', seasonal_periods=4).fit()
fc = model.forecast(8)
fc.index = pd.period_range('2012Q1', periods=8, freq='Q')
fc = fc.round().astype(int)

print('Прогноз продаж VW по кварталам:')
print(fc)
print(f'\nИтого за год — прогноз: 2012 = {fc[:4].sum()}, 2013 = {fc[4:].sum()}')
print('Факт (все модели VW): 2010 = 237, 2011 = 376')

# График: история + прогноз
hist_df = pd.DataFrame({'Квартал': ts.index.to_timestamp(), 'Продажи': ts.values, 'Тип': 'Факт'})
pred_df = pd.DataFrame({'Квартал': fc.index.to_timestamp(), 'Продажи': fc.values, 'Тип': 'Прогноз'})
plot_df = pd.concat([hist_df, pred_df], ignore_index=True)
fig = px.line(plot_df, x='Квартал', y='Продажи', color='Тип', markers=True,
              title='VW: поквартальные продажи — факт 2008–2011 и прогноз 2012–2013')
fig.show()
```
