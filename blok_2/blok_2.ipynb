# Блок 2. Классификация доп.оборудования по группам

Наименования товаров (1574 строки) классифицируются по 5 группам на основе ключевых слов в названии: **коврик, парктроник, сигнализация, защита, камера заднего вида**.

Логика классификации вынесена в `classify_task1.py` / `classify_hybrid.py` (гибрид «правила + LLM с самообучением»), результат сохранён в `task_1_result.xlsx`. Ниже — итоговое распределение и график.

```python
import pandas as pd
import plotly.express as px

# результат классификатора (см. classify_hybrid.py)
df = pd.read_excel('task_1_result.xlsx')
dist = df['группа'].value_counts()
print(dist)
```

```python
print (df['наименование'].value_counts().reset_index())
```

```python
# График: распределение товаров по группам
fig = px.bar(dist.reset_index(), x='группа', y='count', text='count',
             title='Распределение товаров по группам', labels={'count': 'Строк'})
fig.update_traces(textposition='outside')
fig.show()
```

### Вывод

Группировка по назначению (защита / коврик / парктроник / сигнализация / камера) логичнее альтернатив (по типу «товар vs услуга» или по бренду), т.к. отражает потребительские категории и удобна для анализа спроса.
