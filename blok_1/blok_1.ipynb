### Разбор столбцов: кто наши клиенты (таблица сделок)

```python
# --- Категориальные столбцы таблицы СДЕЛОК (реальные клиенты) ---
for col in ['Марка авто', 'Сфера деятельности', 'Цель займа', 'Повторная сделка', 'Стадия сделки']:
    print(f'\n=== {col} ===')
    print(df_sdelka[col].value_counts(dropna=False).head(8))
```

## Вывод: портрет целевой аудитории

Если пройтись по столбцам обеих таблиц, вырисовывается довольно цельный портрет.

Наш типичный клиент — это взрослый человек с небольшим или средним достатком, которому срочно нужны наличные под залог своего автомобиля. По колонкам **«Марка авто», «Год выпуска», «Пробег»** видно, что машина у него недорогая и возрастная: медианный год выпуска — 2005-й (авто около 20 лет), пробег большой (медиана ~275 тыс. км). Марки в основном массовые — Volkswagen, Renault, Lada, Ford; премиум почти не встречается. Для человека это чаще всего единственный ликвидный актив, который можно быстро заложить.

Поле **«Сфера деятельности»** показывает, что это люди рабочих и предпринимательских профессий: торговля, транспорт (перевозки, такси), промышленность, строительство, общепит, сельское хозяйство, попадаются госслужащие. То есть в основном самозанятые и наёмные работники, а не топ-менеджмент.

Столбцы **«Желаемая сумма», «Сумма», «Срок займа»** говорят, что суммы небольшие — около 2 000 рублей (медиана), а срок — примерно год. Это не крупные инвестиции, а закрытие текущих денежных дыр. Подтверждает это и **«Цель займа»**: чаще всего ремонт (авто или квартиры), рефинансирование и закрытие долгов, реже — крупная покупка. Заметная доля «рефинансирования» намекает, что часть клиентов уже закредитована.

Из таблицы лидов (**«Источник», «UTM Source», «UTM Term»**) понятно, как они приходят: в основном через сайт dzengo.by и контекстную рекламу Google/Яндекс, плюс Instagram и финансовый агрегатор MyFin. Поисковые запросы прямые и «горячие»: *«займ онлайн», «микрозайм», «залог авто», «деньги в долг на карту»*. Человек не изучает рынок месяцами — деньги нужны здесь и сейчас.

Наконец, поля **«Повторная сделка»** и **«Повторный лид»** говорят о неплохой возвратности: среди сделок повторных около 44%, среди лидов — примерно четверть. Клиент, которого один раз обслужили, склонен возвращаться (правда, часть «повторности» — это, возможно, не столько лояльность, сколько хроническая нехватка денег).

**Если коротко:** целевая аудитория — это платёжеспособно ограниченные взрослые люди рабочих профессий и малого бизнеса, владеющие недорогим подержанным авто, которым время от времени срочно нужны небольшие суммы наличных на ремонт, закрытие долгов или бытовые нужды. Они ценят скорость, приходят по прямым запросам о займах и часто возвращаются снова.

## Аналитические таблицы (на базе df_all)

1. Конверсия из лида в сделку по источникам.
2. Средняя / желаемая сумма займа и срок.
3. Топ причин отказа и топ успешных сегментов.

> Конверсия считается по `df_all` (лиды + присоединённые сделки). Суммы, срок и сегменты — на уровне сделок (`df_sdelka`, 50 записей), чтобы join по лидам не задваивал строки.

```python
# 1) Конверсия из лида в сделку по источникам (на базе df_all)
conv = df_all.groupby('Источник').agg(
    Лидов=('flag_lid', 'sum'),
    Сделок=('flag_sdelka', 'sum'),
    Успешных=('flag_sdelka_success', 'sum'),
).fillna(0).astype(int)
conv['Конв. в сделку, %'] = (conv['Сделок'] / conv['Лидов'] * 100).round(1)
conv['Конв. в успех, %'] = (conv['Успешных'] / conv['Лидов'] * 100).round(1)
conv.sort_values('Лидов', ascending=False)
```

```python
# График: конверсия в сделку по источникам (каналы с >= 5 лидов)
import plotly.express as px

c = conv[conv['Лидов'] >= 5].reset_index()
fig = px.bar(c, x='Источник', y='Конв. в сделку, %', text='Конв. в сделку, %',
             title='Конверсия в сделку по источникам')
fig.update_traces(textposition='outside')
fig.show()
```

```python
# Тот же график, что вошёл в итоговый отчёт (Рис. 1): лиды + конверсия на двух осях
import matplotlib.pyplot as plt

c2 = conv[conv['Лидов'] >= 5]
fig, ax = plt.subplots(figsize=(8, 4.5))
ax.bar(range(len(c2)), c2['Лидов'], color='#1f77b4', label='Лидов')
ax2 = ax.twinx()
ax2.plot(range(len(c2)), c2['Конв. в сделку, %'], color='#d62728', marker='o')
ax.set_xticks(range(len(c2)))
ax.set_xticklabels(c2.index, rotation=35, ha='right', fontsize=8)
ax.set_ylabel('Лидов', color='#1f77b4')
ax2.set_ylabel('Конверсия в сделку, %', color='#d62728')
ax.set_title('Лиды и конверсия по источникам')
plt.tight_layout()
plt.show()
```

```python
# 2) Средняя / желаемая сумма займа и срок (уровень сделок; все vs успешные)
import re

def _money(x):
    if pd.isna(x):
        return None
    d = re.sub(r'\D', '', str(x).split('.')[0])
    return float(d) if d else None

df_sdelka['Желаемая_сумма_руб'] = df_sdelka['Желаемая сумма (д)'].apply(_money)
usp = df_sdelka['flag_sdelka_success'] == 1

tbl_sum = pd.DataFrame({
    'Все сделки': [df_sdelka['Сумма'].mean(), df_sdelka['Сумма'].median(),
                   df_sdelka['Желаемая_сумма_руб'].mean(), df_sdelka['Желаемая_сумма_руб'].median(),
                   df_sdelka['Срок займа в мес.'].mean(), df_sdelka['Срок займа в мес.'].median()],
    'Успешные': [df_sdelka.loc[usp, 'Сумма'].mean(), df_sdelka.loc[usp, 'Сумма'].median(),
                 df_sdelka.loc[usp, 'Желаемая_сумма_руб'].mean(), df_sdelka.loc[usp, 'Желаемая_сумма_руб'].median(),
                 df_sdelka.loc[usp, 'Срок займа в мес.'].mean(), df_sdelka.loc[usp, 'Срок займа в мес.'].median()],
}, index=['Сумма сделки, ср.', 'Сумма сделки, медиана',
          'Желаемая сумма, ср.', 'Желаемая сумма, медиана',
          'Срок, мес. ср.', 'Срок, мес. медиана']).round(0)
tbl_sum
```

```python
# 3) Топ отказов и топ успешных сегментов (уровень сделок)
from IPython.display import display

print('Топ причин отказа:')
display(df_sdelka[df_sdelka['Стадия сделки'] != 'Сделка успешна']['Стадия сделки']
        .value_counts().to_frame('Кол-во'))

seg_sfera = df_sdelka.groupby('Сфера деятельности').agg(
    Сделок=('flag_sdelka', 'sum'), Успешных=('flag_sdelka_success', 'sum')).astype(int)
seg_sfera['Успех, %'] = (seg_sfera['Успешных'] / seg_sfera['Сделок'] * 100).round(0)
print('Топ успешных сегментов — сфера деятельности:')
display(seg_sfera.sort_values('Успешных', ascending=False))

seg_cel = df_sdelka.groupby('Цель займа').agg(
    Сделок=('flag_sdelka', 'sum'), Успешных=('flag_sdelka_success', 'sum')).astype(int)
seg_cel['Успех, %'] = (seg_cel['Успешных'] / seg_cel['Сделок'] * 100).round(0)
print('Топ успешных сегментов — цель займа:')
display(seg_cel.sort_values('Успешных', ascending=False))
```

```python
# График: доля успешных сделок по сфере деятельности
s = seg_sfera.sort_values('Успех, %').reset_index()
fig = px.bar(s, x='Успех, %', y='Сфера деятельности', orientation='h',
             title='Доля успешных сделок по сфере деятельности')
fig.show()
```

### Что показывают таблицы

**Конверсия по источникам.** Больше всего лидов даёт сайт dzengo.by (184) и Instagram (86), но конверсия в сделку у них невысокая — 3–4%. Самая высокая конверсия у канала «Виджет на сайте / онлайн-чат» — 13% (2 сделки из 15): это тёплый входящий трафик. Звонки и мессенджер-боты (Telegram/Viber) сделок не дали вовсе. Вывод: объём лидов и их качество — разные вещи; онлайн-чат конвертит лучше всего, а по ботам стоит разобраться, почему нет результата.

**Суммы и срок.** Средняя сумма сделки ~3300 руб. при медиане ~1900; желаемая обычно выше фактической (медиана 2000). Успешные сделки в среднем берут на более длинный срок (медиана 24 мес. против 14 по всем) — то есть до сделки чаще доходят те, кто готов к длинному займу.

**Отказы.** Главная причина отказа — «Отказ по КИ» (плохая кредитная история), затем «Двойной залог» и «Отказ клиента». То есть большинство отказов — на стороне скоринга/залога, а не потому что клиент передумал.

**Успешные сегменты.** По сфере деятельности лучше всего конвертируются промышленность (86%), строительство (80%) и транспорт (71%), тогда как торговля — только 30%. По цели займа выше всего успех у «ремонта» и «крупной покупки» (по 80%), ниже — у «рефинансирования/долгов» (60%), что логично: закредитованным чаще отказывают.
